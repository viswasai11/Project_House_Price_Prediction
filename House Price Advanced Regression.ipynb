# Hypothesis Testing

```python
#Test Null Average Salesprice of centralair Yes/No Equal
train_df.SalePrice.groupby(train_df.CentralAir).mean()
```

```python
train_df.SalePrice.groupby(train_df.CentralAir).var()
```

```python
yes=train_df[train_df.CentralAir=='Y']
no=train_df[train_df.CentralAir=='N']
```

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(yes.SalePrice,no.SalePrice,equal_var=False)
# since pvalue=2.280814388186851e-37 is lessthan 0.05,reject null hypothesis
```

Null:There is no significante difference Average Salesprice of centralair.<br>
Alternate:There is a significante difference between Average Salesprice of centralair.<br>

```python
#Test Null Average Salesprice of PavedDrive
train_df.SalePrice.groupby(train_df.PavedDrive).mean()
```

```python
train_df.SalePrice.groupby(train_df.PavedDrive).var()
```

```python
yes1=train_df[train_df.PavedDrive=='Y']
no1=train_df[train_df.PavedDrive=='N']
p1=train_df[train_df.PavedDrive=='P']
```

```python
from scipy.stats import f_oneway
```

```python
f_oneway(yes1.SalePrice,no1.SalePrice,p1.SalePrice)
# since pvalue=1.803568890651533e-18 is lessthan 0.05,reject null hypothesis
```

Null:There is no significante difference between Average Salesprice of PavedDrive.<br>
Alternate:There is a significante difference between Average Salesprice of PavedDrive.<br>

```python
#Test Null No Association between Bldgtype and landslope
pd.crosstab(train_df.BldgType,train_df.LandSlope)
```

```python
from scipy.stats import chi2_contingency
```

Null - There is no Association between BldgType and Landslope<br>
Alt - There is Association between BldgType and Landslope<br>

```python
chi2_contingency(pd.crosstab(train_df.BldgType,train_df.LandSlope))
# since pvalue=0.2629395436820786 is greater than 0.05, failed to reject null hypothesis
```

```python
pd.crosstab(train_df.Utilities,train_df.Street)
```

Null - There is no Association between Utilities and street<br>
Alt - There is Association between Utilities and street<br>

```python
chi2_contingency(pd.crosstab(train_df.Utilities,train_df.Street))
# since pvalue=1.0 is greater than 0.05, failed to reject null hypothesis
```

```python
pd.crosstab(train_df.Alley,train_df.Neighborhood)
```

Null - There is no Association between Alley and Neighorhood<br>
Alt - There is Association between Alley and Neighorhood<br>

```python
chi2_contingency(pd.crosstab(train_df.Alley,train_df.Neighborhood))
# since pvalue=4.141717151623319e-93 is lessthan 0.05,reject null hypothesis
```
